```python
mylist = []
def recursivePermutation(rstringAList, rpermutationList, firstIndex ,lastIndex ):
    
    if firstIndex == lastIndex:
        rpermutationList[lastIndex] = rstringAList[firstIndex]
        if((''.join(rpermutationList).find('13')<0) ):
            print(''.join(rpermutationList))
            mylist.append(''.join(rpermutationList))
            #count+=1
        return 
    else:
        # Here the logic to use i from 0 to length of initial string this makes the code to repeat the chars this is the trick
        for i in range(len(rstringAList)):
            rpermutationList[firstIndex]=rstringAList[i]
            recursivePermutation(rstringAList,rpermutationList,firstIndex+1,lastIndex)

        
def permutationFunction(stringA):
    n = len(stringA)
    # List to iterate and fix chars in, basically the resultant output
    permutationList = ['']*(n)
    # sorted to make them in lexological order
    stringAList = sorted(stringA)
    #print(stringAList)
    recursivePermutation(stringAList,permutationList, 0 , n-1)
```
